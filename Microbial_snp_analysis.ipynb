```python
def max_allele(row):
   if row['#A'] > row['#C'] :
      if row['#A']>row['#G']:
        if row ['#A'] > row['#T']:
         return 'A'
   if row['#C'] > row['#G']:
    if row['#C'] >row['#T']:
        return 'C'
   if row ['#G'] > row['#T']:
        return 'G'
   return 'T'
```

```python
df['max_allele'] = df.apply (lambda row: max_allele (row),axis=1)
```

```python
def max_count(row):
   if row['#A'] > row['#C'] :
      if row['#A']>row['#G']:
        if row ['#A'] > row['#T']:
         return row['#A']
   if row['#C'] > row['#G']:
    if row['#C'] >row['#T']:
        return row['#C']
   if row ['#G'] > row['#T']:
        return row['#G']
   return row['#T']
```

```python
df['max_count'] = df.apply (lambda row: max_count (row),axis=1)
df.head(n=5)
```

```python
def rank_A(row):
  if row['#A'] > row['#C'] & row['#G']:
    if row ['#A'] > row['#T']:
     return 1
  if row['#A'] > row['#C'] & row['#G']:
     if row['#A'] < row['#T']:
        return 2
  if row['#A'] > row['#C'] & row['#T']:
    if row['#A'] < row['#G']:
       return 2
  if row['#A'] > row['#C'] & row['#T']:
     if row['#A'] < row['#G']:
        return 2
  
  return 'unknown'
```

```python
df['rank_A'] = df.apply (lambda row: rank_A (row),axis=1)
df.head(n=5)
```

```python
def min_count(row):
   if row['#T'] !=0:
    if row['#T'] != row['max_count']:
     if row['#T'] > row['#G'] or row['#G'] == row['max_count'] :
        if row['#T']>row['#C'] or row['#C'] == row['max_count'] :
         if row ['#T'] > row['#A'] or row['#A'] == row['max_count']:
            return float(row['#T'])
   if row['#G'] !=0:
    if row['#G'] != row['max_count']:
     if row['#G'] > row['#C']or row['#C']== row['max_count']:
        if row['#G'] > row['#A']or row['#A'] == row['max_count']:
         return float(row['#G'])
   if row['#C'] !=0:
    if row['#C'] != row['max_count']:
     if row ['#C'] > row['#A']or row['#A'] == row['max_count']:
        return float(row['#C'])
   if row['#A'] !=0 & row ['#A'] != row['max_count']:
        return float(row['#A'])
   return 'error'
```

```python
df['min_count'] = df.apply (lambda row: min_count (row),axis=1)
df.head(n=5)
```

```python
def diversity(row):
   if row['#T'] !=0:
    if row['#T'] != row['max_count']:
     if row['#T'] > row['#G'] or row['#G'] == row['max_count'] :
        if row['#T']>row['#C'] or row['#C'] == row['max_count'] :
         if row ['#T'] > row['#A'] or row['#A'] == row['max_count']:
            return float(row['#T'])/702
   if row['#G'] !=0:
    if row['#G'] != row['max_count']:
     if row['#G'] > row['#C']or row['#C']== row['max_count']:
        if row['#G'] > row['#A']or row['#A'] == row['max_count']:
         return float(row['#G'])/702
   if row['#C'] !=0:
    if row['#C'] != row['max_count']:
     if row ['#C'] > row['#A']or row['#A'] == row['max_count']:
        return float(row['#C'])/702
   if row['#A'] !=0 & row ['#A'] != row['max_count']:
        return float(row['#A'])/702
   return 'NaN'
```

```python
df['diversity'] = df.apply (lambda row: diversity (row),axis=1)
df.head(n=5)
```

```python
import numpy as np
from numpy import convolve
import matplotlib.pyplot as plt
 
def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
 
x = df1.Position
y = df1.diversity


yMA = movingaverage(float(y),100)
#print yMA
 
plt.plot(float(x)*[len(float(x)-len(yMA)],yMA)
plt.plot(x,y)
plt.show()
```
